# Transferência de estilo com TensorFlow
Baseado em: [Transferência de estilo neural - Tensor Flow](https://www.tensorflow.org/tutorials/generative/style_transfer)





* **O que é o TensorFlow?**

  TensorFlow é uma plataforma completa, de código aberto, para desenvolvimento de aplicações de machine learning.

* **O que é transferência de estilo?**

  É um processo que usa redes neurais artificiais para aplicar de uma maneira artística o estilo de uma imagem em outra. 







## Pipeline
  

*   Utilizar uma rede neural pré-treinada, do site do TensorFlow Model Hub;
*   Aplicar pré-processamento de imagens para o fazer a transferência de estilo; e
* Aplicar e visualizar a transferência de estilo.



## Importando as bibliotecas e o modelo pré-treinado

```python
import tensorflow_hub as hub 
import tensorflow as tf
from matplotlib import pyplot as plt
import numpy as np
import cv2 #OpenCV
```

#### Modelo pré-treinado

[Hub do TensorFlow - arbitrary-image-stylization-v1-256](https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2)



```python
## Modelo pré-treinado
model = hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
```

## Pré-processamento e upload das imagens

```python
def load_image(img_path):
    img = tf.io.read_file(img_path)
    img = tf.image.decode_image(img, channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32)
    img = img[tf.newaxis, :]
    return img
```

```python
content_image = load_image('papa_dados.png')
style_image = load_image('van_gogh.jpg')
```

## Visualizando a imagem inicial

```python
content_image.shape
```

```python
def imshow(image, title=None):
  if len(image.shape) > 3:
    image = tf.squeeze(image, axis=0)

  plt.imshow(image)
  if title:
    plt.title(title)
```

```python
plt.subplot(1, 2, 1)
imshow(content_image, 'Imagem original')

plt.subplot(1, 2, 2)
imshow(style_image, 'Imagem do estilo')
```

## Imagem estilizada (Van Gogh - Noite Estrelada) 

```python
stylized_image = model(tf.constant(content_image), tf.constant(style_image))[0]
```

```python
plt.subplot(1, 2, 1)
imshow(content_image, 'Imagem original')

plt.subplot(1, 2, 2)
imshow(stylized_image, 'Imagem estilizada')
```

## Imagem estilizada (Carne de jaca)
